4. Write a Python program to read last n lines of a file.


>





```python
def read_last_n_lines(file_name, n):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        last_lines = lines[-n:]

        for line in last_lines:
            print(line, end='')


file_name = "sample.txt"
n = int(input("Enter number of last lines to read: "))

read_last_n_lines(file_name, n)
```

9. Write a Python program to count the frequency of words in a file.


```python

search_word = input("Enter word to search: ").lower()

with open("sample.txt", "r") as file:
    text = file.read().lower()


words = text.split()

count = 0


for i in range(len(words)):
    if words[i] == search_word:
        count += 1

print("Frequency of", search_word, "is:", count)
```

10. Write a Python program to get the file size of a plain file.


```python
with open("sample.txt", "r") as file:
    file.seek(0, 2)   # Move to end of file
    size = file.tell()  # Get position (size)

print("File size:", size, "bytes")
```

13. Write a Python program to read a random line from a file.


```python
import random

with open("sample.txt", "r") as file:
    lines = file.readlines()

# Generate random index
index = random.randint(0, len(lines) - 1)

print("Random line:", lines[index].strip())
```
